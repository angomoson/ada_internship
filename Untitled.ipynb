```python

def convert_mask_values(input_folder, output_folder, num_files=None):
    """
    Convert tif image pixel values:
        0 -> 0
        1 -> 0
        2 -> 1
        3 -> 1

    Parameters
    ----------
    input_folder : str
        Folder containing input .tif images

    output_folder : str
        Folder where converted images will be saved

    num_files : int or None
        Number of files to process.
        If None, process all files.
    """

    input_folder = Path(input_folder)
    output_folder = Path(output_folder)

    # Create output folder if not exists
    output_folder.mkdir(parents=True, exist_ok=True)

    # Get all tif files
    tif_files = sorted(input_folder.glob("*.tif"))

    # Limit number of files if requested
    if num_files is not None:
        tif_files = tif_files[:num_files]

    # Lookup table for ultra-fast mapping
    # index = original pixel value
    # value = mapped pixel value
    lut = np.array([0, 0, 1, 1], dtype=np.uint8)

    for img_path in tif_files:

        # Read image as numpy array
        img_array = np.array(Image.open(img_path))

        # Efficient pixel mapping
        converted = lut[img_array]

        # Save converted image
        output_path = output_folder / img_path.name

        Image.fromarray(converted).save(output_path)

        print(f"Processed: {img_path.name}")


convert_mask_values(
    input_folder="test/target",
    output_folder="test/processed_target",
#     num_files=5
)
```
